```python
#Drop Unnameed Column
data=df.drop("Unnamed: 0",axis=1)
print(data.head(5))
```

```python
plt.figure(figsize=(4,5))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#by analysing that number of female is more than males
gb =df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on their scores
gp =df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
sns.heatmap(gp,annot=True)
plt.title("Relationship btw Parent's Martial Status & Student's Score")
plt.show()
```

```python
#from the above chart there is no neglible impact on parent matrial stauts
#group by on the basis of no. of siblings 
gp =df.groupby("NrSiblings").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gp)
```

```python
#Distribution of Ethnic Groups
groupA=data.loc[(df['EthnicGroup']=="group A")].count()
groupB=data.loc[(df['EthnicGroup']=="group B")].count()
groupC=data.loc[(df['EthnicGroup']=="group C")].count()
groupD=data.loc[(df['EthnicGroup']=="group D")].count()
groupE=data.loc[(df['EthnicGroup']=="group E")].count()
l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
data.plot.pie(y='mlist', labels=data[l], autopct='%1.1f%%')
plt.show()
```
